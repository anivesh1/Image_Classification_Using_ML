```python
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage.transform import resize
```

```python
data_dir = 'D:/Downloads/Img_Rcg_Data'
```

```python
ctgs = ['with_mask' , 'without_mask']
```

```python
flat_data = []
imgs = []
target = []
```

```python
for i in ctgs:
    img_cls = ctgs.index(i)
    path = os.path.join(data_dir, i)
    for img in os.listdir(path):
        img_array = imread(os.path.join(path,img))
        img_resize = resize(img_array , (150,150,3))
        target.append(img_cls)
        flat_data.append(img_resize.flatten())
        imgs.append(img_resize)
```

```python
flat_data
```

```python
target = np.array(target)
```

```python
target
```

```python
target.shape
```

```python
np.unique(target , return_counts = True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(flat_data , target , test_size = 0.2  , random_state = 0)
```

### USING SVC

```python
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = [{'C': (1,10,100,500,1000) } ,
          {'C': (1,10,100,500,1000) , 'gamma':(0.001,0.01,0.1,0.5,1)}]
```

```python
clf = SVC(probability = True)
```

```python
clf_grid = GridSearchCV(clf , params , n_jobs = -1, cv = 3 , verbose = 1, scoring = "accuracy")
```

```python
clf_grid.fit(X_train , y_train)
```

```python
clf_grid.best_params_
```

```python
clfb = clf_grid.best_estimator_
```

```python
clfb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test , clfb.predict(X_test))
```
